```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/bengaluru_house_prices.csv')
df.shape
```

```python
df.head()
```

```python
df2=df.drop(['area_type','society','balcony','availability','bath'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3.head()
```

```python
df3['size'].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head(10)
```

```python
df3['total_sqft'].unique()
```

```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
df4 = df3[df3['total_sqft'].apply(is_float)]
df4.drop(columns=['size'], inplace=True)
df4.head()
```

```python
df5 = df4.copy()
df5['total_sqft'] = pd.to_numeric(df5['total_sqft'], errors='coerce' )
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```
